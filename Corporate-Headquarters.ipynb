## All 15 locations:

```python
import vincent, json
import folium
import os #import to save map to .html file 
from folium.plugins import MarkerCluster

# instantiate folium map with starting map point as location
map1 = folium.Map(location=(ch_df['latitude'].mean(), ch_df['longitude'].mean()), tiles="OpenStreetMap", zoom_start=5)


marker_cluster = MarkerCluster().add_to(map1)

#iterate through table rows
for index, row in ch_df.sample(15).iterrows():
    #set to latitude/longitude coordinates
    coord = [row['latitude'], row['longitude']]

    #add marker to map
    folium.Marker(
        location=coord,
        popup="Corprate Headquarters",
        icon=folium.Icon(color='black',icon='info-sign'),
    ).add_to(marker_cluster)

#save map as .html file
map1.save(os.path.join('data', 'map.html'))

map1
```
